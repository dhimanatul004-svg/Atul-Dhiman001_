```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Sample Dataset
data = {
    "OrderID": [101,102,103,104,105,106,107,108,109,110],
    "Region": ["North","South","East","West","North","South","East","West","North","South"],
    "Product": ["Laptop","Mouse","Keyboard","Laptop","Mouse","Keyboard","Laptop","Mouse","Keyboard","Laptop"],
    "Quantity": [2,5,3,1,4,6,2,7,5,3],
    "Price": [50000,500,1500,52000,600,1400,51000,550,1600,53000]
}

df = pd.DataFrame(data)

# Total Sales Column
df["Total_Sales"] = df["Quantity"] * df["Price"]

print(df)
```

```python
# When and Why to Use Different Charts
# 1. Bar Chart

# When to Use:
# When comparing values across categories.
# Example: Region-wise sales, Product-wise revenue.

# Why to Use:
# It clearly shows differences between groups.
# Easy to identify highest and lowest performers.

# Best For:
# Categorical data comparison.

# 2. Line Chart

# When to Use:
# When analyzing trends over time.
# Example: Monthly sales trend, yearly growth.

# Why to Use:
# Shows patterns, growth, or decline.
# Best for time-series data.

# Best For:
# Continuous data over time.

# 3. Pie Chart

# When to Use:
# When showing percentage contribution.
# Example: Product contribution to total sales.

# Why to Use:
# Displays proportion of each category.
# Good for part-to-whole comparison.

# Best For:
# Showing composition (limited categories).

# 4. Histogram

# When to Use:
# When understanding data distribution.
# Example: Sales frequency distribution.

# Why to Use:
# Shows how data is spread.
# Helps identify skewness and frequency patterns.

# Best For:
# Continuous numerical data distribution.

# 5. Scatter Plot

# When to Use:
# When analyzing relationship between two numeric variables.
# Example: Quantity vs Price.

# Why to Use:
# Shows correlation or pattern between variables.
# Helps detect trends and clusters.

# Best For:
# Relationship analysis.

# 6. Box Plot

# When to Use:
# When analyzing distribution with outliers.
# Example: Region-wise sales variation.

# Why to Use:
# Shows median, quartiles, and outliers.
# Helps compare spread across categories.

# Best For:
# Detecting variation and outliers.

# Simple Interview Summary (Very Important)

# Use Bar Chart for comparison.
# Use Line Chart for trends over time.
# Use Pie Chart for percentage contribution.
# Use Histogram for distribution analysis.
# Use Scatter Plot for relationship between variables.
# Use Box Plot for spread and outlier detection.
```

```python
print("Mean Sales:", np.mean(df["Total_Sales"]))
print("Median Sales:", np.median(df["Total_Sales"]))
print("Standard Deviation:", np.std(df["Total_Sales"]))##agr mean se value kam aati hai iska matlb hai ki data stabel hai {standard deviation to check if data is stable or not (unstable)} 
print("Minimum Sales:", np.min(df["Total_Sales"]))
print("Maximum Sales:", np.max(df["Total_Sales"]))
```

```python
region_sales = df.groupby("Region")["Total_Sales"].sum()##similar data ko  group krne ke liye groupby 
print(region_sales)
```

```python
region_sales.plot(kind="bar")
plt.title("Region Wise Total Sales")
plt.xlabel("Region")
plt.ylabel("Total Sales")
plt.show()
```

```python
plt.plot(df["OrderID"], df["Total_Sales"])
plt.title("Sales Trend")
plt.xlabel("OrderID")
plt.ylabel("Total Sales")
plt.show()
```

```python
product_sales = df.groupby("Product")["Total_Sales"].sum()

plt.pie(product_sales, labels=product_sales.index, autopct="%1.1f%%")
plt.title("Product Wise Sales Distribution")
plt.show()
```

```python
plt.hist(df["Total_Sales"], bins=5)
plt.title("Sales Distribution")
plt.xlabel("Total Sales")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.scatter(df["Quantity"], df["Price"])
plt.title("Quantity vs Price")
plt.xlabel("Quantity")
plt.ylabel("Price")
plt.show()
```

```python
print("25th Percentile:", np.percentile(df["Total_Sales"], 25))##if 0 to 1 then quentile ,0 to 100 percentile
print("50th Percentile:", np.percentile(df["Total_Sales"], 50))
print("75th Percentile:", np.percentile(df["Total_Sales"], 75))
```

```python
df["Cumulative_Sales"] = np.cumsum(df["Total_Sales"])##cumulative sum (cumsum())(1,2,3)={(1,3,6)cumsum}
print(df[["OrderID","Cumulative_Sales"]])
```

```python
plt.boxplot(df["Total_Sales"])
plt.title("Box Plot of Total Sales")
plt.ylabel("Total Sales")
plt.show()
```

```python
north = df[df["Region"]=="North"]["Total_Sales"]
south = df[df["Region"]=="South"]["Total_Sales"]
east = df[df["Region"]=="East"]["Total_Sales"]
west = df[df["Region"]=="West"]["Total_Sales"]

plt.boxplot([north, south, east, west], labels=["North","South","East","West"])
plt.title("Region Wise Sales Distribution")
plt.ylabel("Total Sales")
plt.show()
```

```python
df.boxplot(column="Total_Sales", by="Region")
plt.title("Region Wise Boxplot")
plt.suptitle("")
plt.show()
```
